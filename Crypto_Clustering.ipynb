```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
df_normalized = scaler.fit_transform(df_market_data)

# Display first five rows of the scaled data
df_normalized[0:5]
```

```python
# Create a DataFrame with the scaled data
df_normalized_scaled = pd.DataFrame(
    df_normalized,
    columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]
)

# Copy the crypto names from the original data
df_normalized_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_normalized_scaled = df_normalized_scaled.set_index("coin_id")

# Display sample data
df_normalized_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_normalized_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer: 3

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=3, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_normalized_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeans_predictions = model.predict(df_normalized_scaled)

# Print the resulting array of cluster values.
print(kmeans_predictions)
```

```python
# Create a copy of the DataFrame
predictions_df = df_normalized_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
predictions_df["kmeans-segments"] = kmeans_predictions

# Display sample data
predictions_df[['kmeans-segments']].head(5)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="kmeans-segments",
    hover_cols = ["coin_id"] 
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
predictions_pca = pca.fit_transform(predictions_df)
# View the first five rows of the DataFrame. 
predictions_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** About 88% of the total variance is condensed into the 3 PCA variables.

```python
# Create a new DataFrame with the PCA data.
predictions_pca_df = pd.DataFrame(
    predictions_pca,
    columns=["PCA1", "PCA2", "PCA3"]
)

# Copy the crypto names from the original data
predictions_pca_df["coin_id"] = df_market_data.index

# Set the coinid column as index
predictions_pca_df = predictions_pca_df.set_index("coin_id")

# Display sample data
predictions_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(predictions_pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
# Review the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 3


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=3, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(predictions_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_3 = model.predict(predictions_pca_df)
# Print the resulting array of cluster values.
print(k_3)
```

```python
# Create a copy of the DataFrame with the PCA data
predictions_pca_df = predictions_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
predictions_pca_df["predicted_clusters"] = k_3

# Display sample data
print(predictions_pca_df)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
predictions_pca_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted_clusters",
    hover_cols = ["coin_id"] 
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves

# Create a list with the number of k-values from 1 to 11
k_original = list(range(1, 12))

# Compute inertia values for the original data
inertia_original = []

for i in k_original:
    k_model_original = KMeans(n_clusters=i, random_state=0)
    k_model_original.fit(df_normalized_scaled)
    inertia_original.append(k_model_original.inertia_)

# PCA-optimized data
# Repeat the steps for PCA-optimized clustering and obtain the optimal k value

# Composite plot
original_elbow_data = {"k": k_original, "inertia": inertia_original}
df_original_elbow = pd.DataFrame(original_elbow_data)

pca_elbow_data = {"k": k, "inertia": inertia}
df_pca_elbow = pd.DataFrame(pca_elbow_data)

# Plot the Elbow curves for both cases on the same chart
original_elbow_curve = df_original_elbow.hvplot.line(
    x="k",
    y="inertia",
    line_color="red",
    label="Original Data"
)

pca_elbow_curve = df_pca_elbow.hvplot.line(
    x="k",
    y="inertia",
    line_color="blue",
    label="PCA-Optimized Data"
)

composite_plot = (original_elbow_curve * pca_elbow_curve).opts(legend_position='top_left')

composite_plot
```

```python
# Composite plot to contrast the clusters

# Create a scatter plot for original data clusters
original_scatter = predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="kmeans-segments",
    hover_cols = ["coin_id"], 
    title="Original Data Clusters"
)

# Create a scatter plot for PCA-optimized data clusters
pca_scatter = predictions_pca_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted_clusters",
    hover_cols = ["coin_id"],
    title="PCA-Optimized Data Clusters"
)

# Display the scatter plots
original_scatter + pca_scatter

```
